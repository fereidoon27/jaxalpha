```python
import zipfile
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the folder to zip
folder_to_zip = '/content/ParallelFold/output'

# Define the output zip file name and path (in Google Drive)
output_filename = '/content/drive/My Drive/GA_OUTPUT_COLAB.zip'

# Create the zip file
with zipfile.ZipFile(output_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(folder_to_zip):
        for file in files:
            zipf.write(os.path.join(root, file),
                       os.path.relpath(os.path.join(root, file),
                                       os.path.join(folder_to_zip, '..')))

print(f"Zip file created and saved to Google Drive: {output_filename}")
```
